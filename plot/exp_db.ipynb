```python
import plot
options = plot.Options(
    save=False,
    plot_nothing=True,
    plot_db=True,
    plot_pressure=True,
    print_params=False,
    print_pressure_values=True,
    db_mean_interval=2,
    plot_all_pressure=True)
exp_dir='exp_db'
exp_bs = [4,8,16,32,64,128,256,512]
```

## YCSB Workload A

```python
options.file_start_time[f'{exp_dir}/ycsb_wa.out'] = 30
options.db_xlim = [-0.05,     60.05]
options.db_ylim = [ 0   ,  72000   ]
plot.plotFiles(
    [f'{exp_dir}/ycsb_wa.out'] + [f'{exp_dir}/ycsb_wa,at3_bs{bs}_directio.out' for bs in exp_bs],
    options, allfiles=f'{exp_dir}/ycsb_wa,at3_bsX_directio')
```
