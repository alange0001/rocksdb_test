```python
import plot
options = plot.Options(
    save=False,
    plot_nothing=True,
    plot_db=True,
    plot_pressure=True,
    print_params=False,
    db_mean_interval=2)
```

## YCSB Workload A

```python
options.file_start_time['exp_db/ycsb_wa.out'] = 30
options.db_xlim = [-0.05,     60.05]
options.db_ylim = [ 0   ,  72000   ]
a = plot.AllFiles('exp_db/ycsb_wa,at3_bsX_directio')
plot.plotFiles([
    'exp_db/ycsb_wa.out',
    'exp_db/ycsb_wa,at3_bs4_directio.out',
    'exp_db/ycsb_wa,at3_bs8_directio.out',
    'exp_db/ycsb_wa,at3_bs16_directio.out',
    'exp_db/ycsb_wa,at3_bs32_directio.out',
    'exp_db/ycsb_wa,at3_bs64_directio.out',
    'exp_db/ycsb_wa,at3_bs128_directio.out',
    'exp_db/ycsb_wa,at3_bs256_directio.out',
    'exp_db/ycsb_wa,at3_bs512_directio.out',
], options, a)
print("######################################################")
print("dbmean:")
a.plot_dbmean()
```
