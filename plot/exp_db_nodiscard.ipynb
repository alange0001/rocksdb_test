```python
import plot
options = plot.Options(
    save=False,
    plot_nothing=True,
    plot_db=True,
    plot_pressure=True,
    print_params=False,
    db_mean_interval=2)

exp_dir='exp_db_nodiscard'
exp_bs = ['4', '8', '16', '32', '64', '128', '256', '512']
```

## YCSB Workload A

```python
files = [f'{exp_dir}/ycsb_workloada.out'] + \
        [f'{exp_dir}/ycsb_workloada,at3_bs{bs}_directio.out' for bs in exp_bs]
a = plot.AllFiles('exp_db/ycsb_workloada,at3_bsX_directio')

options.db_xlim = [-0.05,     60.05]
options.db_ylim = [ 0   ,  40000   ]

plot.plotFiles(files, options, a)
print("######################################################")
print("dbmean:")
a.plot_dbmean()
```
